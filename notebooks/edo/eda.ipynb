```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../data/interim/interim_data.csv", low_memory=False)
df.head()
```

```python
df.dropna(axis=0)
```

```python
df.fillna(0, inplace=True)
```

```python
object_type = df.select_dtypes(include=['object']).columns.to_list()
object_type
```

```python
# take care of null values
def remove_groups_with_only_nan(df, column):
    mask = df.groupby('match_id')[column].transform(lambda x: (x == "0").all())
    return df[~mask]
for col in object_type:
    df_filtered = remove_groups_with_only_nan(df, col)
df_filtered 
```
